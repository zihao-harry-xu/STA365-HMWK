## Part C (II): Full Conditional [Posterior] Distributions

### 1. What is $p(\theta|\tau, \mathbf{x})$?<br>This is a so-called a *full conditional* [posterior distribution]

#### *This is actually just asking for $p(\theta|\tau, \mathbf{x}) \propto p(\textbf{x}, \theta|\tau) \propto p(\textbf{x},\theta,\tau)$*<br>which is just $p(\textbf{x},\theta,\tau)$ viewed only as a function of $\theta$ with $\textbf{x}$ given<br>*(which means $\textbf{x}$ and $\tau$ are known because they've been observed)*

### 2. What is $p(\theta,\tau | \mathbf{x})$ this if all normalizing constants are removed if ONLY $\tau$ is was a random variable?

### That is, what is the *full conditional* [posterior distribution] $p(\tau|\theta, \mathbf{x})$?

$\LARGE

## Part C (III): Gibbs Sampling

The canonical example of ***Gibbs sampling*** is the ***normal-gamma*** model and here's a simplified version with $p(\theta,\tau) \propto 1$

$$\begin{align*}
p\left(x|\theta,\tau=\frac{1}{\sigma^2}\right) = {} & \prod_{i=1}^n \sqrt{\frac{\tau}{2\pi}} e^{-\frac{\tau}{2}(x_i-\theta)^2}\\
p(\theta,\tau) \propto {} & 1 \\
p(\theta,\tau| x) \propto p(\theta,\tau, x) \propto {} &  \tau^{n/2} e^{-\frac{\tau}{2} \sum_{i=1}^n (x_i-\theta)^2} \\
p(\tau|\theta, x) = {} & \text{Gamma}\left(\alpha = \frac{n}{2}+1, \beta= \frac{1}{2} \sum_{i=1}^n (x_i-\theta)^2 \right)\\
p(\theta| \tau, x)  \propto {} & e^{-\frac{\tau}{2} \sum_{i=1}^n \theta^2 - 2\theta x_i} =  e^{-\frac{\tau}{2} \left(n\theta^2 - 2\theta \sum_{i=1}^n x_i\right)} = e^{-\frac{\tau n }{2} \left(\theta^2 - 2\theta \bar x\right)}\\
  \propto {} & e^{-\frac{n \tau}{2} (\theta - \bar x)^2} \\
p(\theta| \sigma^2 = \tau^{-1}, x) = {} & N\left(E[\theta] = \bar x, Var(\theta) = \frac{\sigma^2}{n}\right)
\end{align*}$$

Derive the *full conditional distributions* $p(\tau|\theta, \mathbf{x})$ and $p(\theta|\tau, \mathbf{x})$ if $p(\theta)$ is $N(\theta_0, \tau^2_0)$ and $p(\tau) \propto \tau^{\frac{\alpha}{2}-1} e^{- \frac{\lambda}{2} \tau}$. 

*Explain how to equate $\alpha$ and $\lambda$ with controlling the amount and nature of information the prior contributes to the posterior. How can we interpret these as controlling a "number of data points" contributed by the prior relative to the number of data points contributed to the posterior through the likelihoood?*

Create an updated version of the *Gibbs sampler* for $N(\theta_0, \tau^2_0)$ and $\textrm{gamma}(\textrm{shape}=\frac{\alpha}{2}, \textrm{rate}=\frac{\lambda}{2})$ priors for $p(\theta)$ and $p(\tau)$ respectively.

```python
# QUESTION 1 ANSWER

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

np.random.seed(42)
n = 30
m, s = 0, 1 
x = stats.norm(loc=m, scale=s).rvs(size=n)


C, G = 2, 10000
theta, tau = np.zeros([C, G]), np.zeros([C, G])

# Initialize theta and tau
theta[:, 0] = np.mean(x)
tau[:, 0] = 1 / np.var(x)

# Gibbs Sampler
for c in range(C):
    for g in range(1, G):
        # Sample tau from Gamma
        alpha = n / 2
        beta = 0.5 * np.sum((x - theta[c, g-1])**2)
        tau[c, g] = stats.gamma(a=alpha, scale=1/beta).rvs()

        # Sample theta from Normal
        theta_mean = np.mean(x)
        theta_var = (1 / (tau[c, g] * n))
        theta[c, g] = stats.norm(loc=theta_mean, scale=np.sqrt(theta_var)).rvs()

# Plotting Burn-in and Convergence
fig, ax = plt.subplots(1, 4, figsize=(14, 3))
ax[0].set_title("$\\theta$ Burn-In")
ax[1].set_title("$\\tau$ Burn-In")
ax[2].set_title("$\\theta$ Converged Markov Chain\nStationary Distribution Samples")
ax[3].set_title("$\\tau$ Converged Markov Chain\nStationary Distribution Samples")

burn = 500 
demo = 120  

for c in range(C):
    ax[0].plot(theta[c, :burn], label=f"$\\theta$ Chain {c}")
    ax[1].plot(tau[c, :burn], label=f"$\\tau$ Chain {c}")
    ax[2].plot(np.arange(burn, demo + burn, dtype=int), theta[c, burn:demo+burn], label=f"$\\theta$ Chain {c}")
    ax[3].plot(np.arange(burn, demo + burn, dtype=int), tau[c, burn:demo+burn], label=f"$\\tau$ Chain {c}")

ax[0].legend()
ax[1].legend()
ax[2].legend()
ax[3].legend()
plt.show()

# Bayesian credible intervals
theta_samples = theta[:, burn:].flatten()  # Remove burn-in
tau_samples = tau[:, burn:].flatten()  # Remove burn-in

credible_interval_theta = np.quantile(theta_samples, [0.025, 0.975])
credible_interval_tau = np.quantile(tau_samples, [0.025, 0.975])

print(f"95% Bayesian Credible Interval for θ: {credible_interval_theta}")
print(f"95% Bayesian Credible Interval for τ: {credible_interval_tau}")
```

```python
# Question 2, ### PLEASE REFER TO HANDWRITTEN PDF FILE IN FOLDER FOR DERIVATION AND EXPLANATION SECTION ###


# Creating updated version of gibbs sampler
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(42)
n = 30
m, s = 0, 1  # True mean and std deviation
x = stats.norm(loc=m, scale=s).rvs(size=n)

# Define Gibbs Sampling Function
def gibbs_sampler(theta_0, tau_0, q, lambda_, C=2, G=10000):
    theta, tau = np.zeros([C, G]), np.zeros([C, G])

    # Initialize theta and tau
    theta[:, 0] = np.mean(x)
    tau[:, 0] = 1 / np.var(x)

    # Gibbs Sampler
    for c in range(C):
        for g in range(1, G):
            # Sample tau from Gamma
            alpha = (n + q) / 2
            beta = (np.sum((x - theta[c, g-1])**2) + lambda_) / 2
            tau[c, g] = stats.gamma(a=alpha, scale=1/beta).rvs()

            # Sample theta from Normal
            theta_mean = (n * np.mean(x) + tau_0 * theta_0) / (n + tau_0)
            theta_var = 1 / ((n + tau_0) * tau[c, g])
            theta[c, g] = stats.norm(loc=theta_mean, scale=np.sqrt(theta_var)).rvs()

    return theta, tau

# Gibbs Sampling with Two Different Priors
# First run (original priors)
theta_0_1, tau_0_1, q_1, lambda_1 = 0, 1, 2, 2
theta_1, tau_1 = gibbs_sampler(theta_0_1, tau_0_1, q_1, lambda_1)

# modified priors
theta_0_2, tau_0_2, q_2, lambda_2 = 5, 10, 10, 10  # More informative priors
theta_2, tau_2 = gibbs_sampler(theta_0_2, tau_0_2, q_2, lambda_2)

# Burn-in and Convergence for Both Cases
burn = 500  # Burn-in period
demo = 120  # Show first 120 post-burn-in samples

fig, ax = plt.subplots(2, 4, figsize=(14, 6))  # Two rows, four columns
titles = ["$\\theta$ Burn-In", "$\\tau$ Burn-In", 
          "$\\theta$ Converged Markov Chain", "$\\tau$ Converged Markov Chain"]

for i, (theta, tau, title_suffix) in enumerate(zip([theta_1, theta_2], [tau_1, tau_2], [" (Original Priors)", " (Modified Priors)"])):
    for c in range(C):
        ax[i, 0].plot(theta[c, :burn], label=f"$\\theta$ Chain {c}")
        ax[i, 1].plot(tau[c, :burn], label=f"$\\tau$ Chain {c}")
        ax[i, 2].plot(np.arange(burn, demo + burn, dtype=int), theta[c, burn:demo+burn], label=f"$\\theta$ Chain {c}")
        ax[i, 3].plot(np.arange(burn, demo + burn, dtype=int), tau[c, burn:demo+burn], label=f"$\\tau$ Chain {c}")

    for j in range(4):
        ax[i, j].set_title(titles[j] + title_suffix)
        ax[i, j].legend()

plt.tight_layout()
plt.show()

# Bayesian credible intervals
theta_samples_1 = theta_1[:, burn:].flatten()
theta_samples_2 = theta_2[:, burn:].flatten()
tau_samples_1 = tau_1[:, burn:].flatten()
tau_samples_2 = tau_2[:, burn:].flatten()

credible_interval_1 = np.quantile(theta_samples_1, [0.025, 0.975])
credible_interval_2 = np.quantile(theta_samples_2, [0.025, 0.975])
credible_interval_tau_1 = np.quantile(tau_samples_1, [0.025, 0.975])
credible_interval_tau_2 = np.quantile(tau_samples_2, [0.025, 0.975])

print(f"95% Bayesian Credible Interval for θ (Original Priors): {credible_interval_1}")
print(f"95% Bayesian Credible Interval for θ (Modified Priors): {credible_interval_2}")
print(f"95% Bayesian Credible Interval for τ (Original Priors): {credible_interval_tau_1}")
print(f"95% Bayesian Credible Interval for τ (Modified Priors): {credible_interval_tau_2}")
```

  &={} p(\boldsymbol{\theta}|\textbf{x})
\end{align*}}
$    



We can see that when we have ~~**conjugate** priors (where the prior and posterior have the same form)~~ <u>**just any (propor) prior**</u>, the denominator in Bayes rule amounts to any normalizing constants in the likelihood and prior (that allow for the densities to integrate to 1 times the inverse of the equivalent normalizing constant of the posterior. Once we plug this back in to Bayes formula and cancel and rearrange terms, this amounts to multiplying the normalizing constant of the posterior with the portion of the posterior that relies on the parameter. This sort of derivation isn't unique to conjugate priors; as long as we can recognize the family in which the product of the prior and likelihood belong to, the integral in the denominator can be solved by simplifying the term in the integral to the precise density (which occurs by moving normalizing constants outside of the integral, and introducing the normalizing constant for that new density). This is the technique we used to solve the integral above.

As a result, we can use a proportionality argument to skip all of these mathematical derivations. If we can recognize the functional form of the product between the likelihood and prior as being in the family of another probability distribution, then that is equivalent to moving around these normalizing constants to simplify integrals.
