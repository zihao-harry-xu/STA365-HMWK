### Q3: Slice Sampling

*First explain how the Markov algorithm of slice sampling as given below works. Then explain the steps by which slice sampling could be used in place of a Metropolis-Hasting step in a Metropolis within Gibbs algorithm where the full conditionals are only known up to a normalizing constant. In your explanation clarify what the curve that we're sampling beneath is, and what the initial value and steps are to create the draw for Gibbs sampling.*

```python
### Answer to Q2 ###
import pymc as pm
import numpy as np
import arviz as az
import matplotlib.pyplot as plt

# Sample Data
np.random.seed(42)
x = np.random.normal(loc=0, scale=1, size=30)

# 1. Normal-Gamma Model
with pm.Model() as model_normal_gamma:
    theta = pm.Normal("theta", mu=0, sigma=1)  # Normal prior
    tau = pm.Gamma("tau", alpha=2, beta=2)     # Gamma prior
    
    # Likelihood (Using sigma)
    likelihood = pm.Normal("likelihood", mu=theta, sigma=1/pm.math.sqrt(tau), observed=x)

    # Sampling
    trace_normal_gamma = pm.sample(2000, tune=1000, chains=2, random_seed=42)

# Diagnostics
print(az.summary(trace_normal_gamma))
az.plot_trace(trace_normal_gamma)
plt.show()


# 2. Uniform-LogNormal Model
with pm.Model() as model_uniform_lognormal:
    theta = pm.Uniform("theta", lower=-5, upper=5)  # Uniform prior
    tau = pm.LogNormal("tau", mu=0, sigma=1)        # Log-Normal prior
    
    likelihood = pm.Normal("likelihood", mu=theta, sigma=1/pm.math.sqrt(tau), observed=x)

    # Sampling
    trace_uniform_lognormal = pm.sample(2000, tune=1000, chains=2, random_seed=42)

# Diagnostics
print(az.summary(trace_uniform_lognormal))
az.plot_energy(trace_uniform_lognormal)
plt.show()


# 3. Laplace-InverseGamma Model
with pm.Model() as model_laplace_invgamma:
    theta = pm.Laplace("theta", mu=0, b=1)            # Laplace prior
    tau = pm.InverseGamma("tau", alpha=3, beta=2)      # Inverse-Gamma prior
    
    likelihood = pm.Normal("likelihood", mu=theta, sigma=1/pm.math.sqrt(tau), observed=x)

    # Sampling
    trace_laplace_invgamma = pm.sample(2000, tune=1000, chains=2, random_seed=42)

# Diagnostics
print(az.summary(trace_laplace_invgamma))
az.plot_forest(trace_laplace_invgamma, combined=True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats  # Import scipy.stats to fix the error
```

```python

def slice_f_at_y(f, x, y, x_grid=np.linspace(0,1,51)):
    
    # find interval of grid points where f(x_grid) > y 
    # then extend the enterval so endpoints f(a)<y and f(b)<y 
    x_grid_delta = x_grid[1]-x_grid[0]
    a,b = x_grid[f(x_grid)>y][[0,-1]]+[-x_grid_delta,x_grid_delta]
    # a,b = x_grid[0,-1]  # make the interval all of x_grid
    
    x_ = a + stats.uniform().rvs()*(b-a)    
    if f(x_)>y:
        return x_,1  # in 1 try if f(x_)>y and "x_ is under f" 
    elif x_ < x:  # or if "x_ was above f on the left side of the interval" 
        x_l,x_r = x_,b
    else:
        x_l,x_r = a,x_  # or if "x_ was above f on the right side of the interval"
    return slice_f_at_y_(f, x, y, x_l, x_r, tot = 2)  # try again with a reduced interval

def slice_f_at_y_(f, x, y, x_l=0, x_r=1, tot=1):
    
    x_ = x_l + stats.uniform().rvs()*(x_r-x_l)
    
    if f(x_)>y:
        return x_,tot
    elif x_ < x:
        x_l = x_
    else:
        x_r = x_
    return slice_f_at_y_(f, x, y, x_l, x_r, tot = tot+1)
```

```python

x_grid = np.linspace(0,1,1000)
f = lambda x: stats.beta(1.5,3).pdf(x)
plt.plot(x_grid, f(x_grid))
    
m = 1000
x = np.zeros([m+1,3])
x[:2,0] = 0.25

plot_trace = 10
for t in range(1,m):
    
    x[t,1] = stats.uniform().rvs()*f(x[t,0])
    if t < plot_trace:
        plt.plot([x[t,0]]*2, [x[t-1,1],x[t,1]], 'k')
    
    x[t+1,0],x[t+1,2] = slice_f_at_y(f, x[t,0], x[t,1])
    if t < plot_trace:
        if t==1:
            plt.plot([x[t,0], x[t+1,0]], [x[t,1]]*2, 'k--', label=str(plot_trace)+ " iterations")
        plt.plot([x[t,0], x[t+1,0]], [x[t,1]]*2, 'k--')
    
plt.hist(x[:,0], density=True, label=str(m)+" iterations\n x values only")
plt.legend();
```

##### Question 3 ANSWER
The slice sampling algorithm works by generating samples from a target distribution by uniformly sampling under its porbability density curve. It uses auxiliary variables to create a "slice" where the density exceeds a randomly chosen height. There are 3 core components, the first being the target curive which is the unnmorlized PDF, f(x) that we want to sample from which in this case is the beta distribution. The next is the slice which is the region where f(x) >= y and where y is a uniform random variable between 0 and f(x). The algorithm also dynamically adjusts the horizontal interval to ensure all points in the slice are captured. In place of a Metropolis-Hasting step in a Metropolis within Gibbs algorithm, slice sampling can avoid explicit normalization by exploiting proportionality. For each Gibbs Step, slice sampling can update the parameter theta with full conditional. We first define the Gibbs sampling framework which requires updating each parameter sequentially from its full conditional distribution. Second, slice sampling works by introducing an auxiliary variable y which represents a horizontal slice bneneath the probability curve f(x). Then, a vertical slice height y can be used to define the horizontal slice beneath the curve. An interval is then located where f(x) > y which can be done by expanding outwards until both ends of the interval, a and b satisfy f(a) < y and f(b) < y. Once obtained, a new x_i can be obtained and accepted as the next Gibbs sample. Therefore, Slice Sampling replaces Metropolis-Hastings by directly defining an interval base don the full coinditional and ensuring the sample x' is always accepted.
